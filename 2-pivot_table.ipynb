```python
# Importando as bibliotecas necessárias

import pandas as pd
```

```python
# 1 - Importando os dados 

df = pd.read_excel('data/VendaCarros.xlsx')
df
```

```python
# 2 - Selecionando as colunas específicas

df_colunas = df[['Fabricante', 'ValorVenda', 'Ano']]
df_colunas
```

```python
# 3 - Criando a tabela pivô

df_pivot = pd.pivot_table(index='Ano', columns='Fabricante', values='ValorVenda', aggfunc='sum', data=df)
df_pivot
```

```python
# 4 - Exportando tabela pivô para um arquivo excel

df_pivot.to_excel('data/VendaCarrosPivot.xlsx', sheet_name='Relatorio')
```
